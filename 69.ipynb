```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
```

```python
dataset_name = "caltech101"
dataset = tfds.load(dataset_name, split=tfds.Split.TRAIN)
```

```python
for example in dataset.take(5):
    image = example["image"]
    label = example["label"]

# Візуалізація зображення
plt.imshow(image.numpy())
plt.axis("off")
plt.title("Оригінальне зображення")
plt.show()
```

```python
# Дзеркальне відображення
flipped_image = tf.image.flip_left_right(image)

# Зміна насиченості кольорів
saturated_image = tf.image.adjust_saturation(image, saturation_factor=2.0)

# Відображення в сірому
grayscale_image = tf.image.rgb_to_grayscale(image)

# Поворот на 90 градусів
rotated_image = tf.image.rot90(image)

# Обрізка по краях
cropped_image = tf.image.crop_to_bounding_box(
    image,
    offset_height=10,
    offset_width=20,
    target_height=150,
    target_width=200
)

# Зміна яскравості
brightness_adjusted_image = tf.image.adjust_brightness(image, delta=0.2)
```

```python
# Візуалізація зображень після аугментаційних перетворень
fig, axes = plt.subplots(2, 3, figsize=(10, 8))
axes[0, 0].imshow(flipped_image.numpy())
axes[0, 0].set_title("Дзеркальне відображення")
axes[0, 0].axis("off")

axes[0, 1].imshow(saturated_image.numpy())
axes[0, 1].set_title("Зміна насиченості кольорів")
axes[0, 1].axis("off")

axes[0, 2].imshow(grayscale_image.numpy().squeeze(), cmap="gray")
axes[0, 2].set_title("Відображення в сірому")
axes[0, 2].axis("off")

axes[1, 0].imshow(rotated_image.numpy())
axes[1, 0].set_title("Поворот на 90 градусів")
axes[1, 0].axis("off")

axes[1, 1].imshow(cropped_image.numpy())
axes[1, 1].set_title("Обрізка по краях")
axes[1, 1].axis("off")

axes[1, 2].imshow(brightness_adjusted_image.numpy())
axes[1, 2].set_title("Зміна яскравості")
axes[1, 2].axis("off")

plt.tight_layout()
plt.show()
```
